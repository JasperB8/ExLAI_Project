# Logistic Regression Model

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Load our dataset and split into specific features and target
df = pd.read_csv("persuade_2.0_human_scores_demo_id_github.csv")
df.keys()
```

```python
targets = df['holistic_essay_score']
data = df[['grade_level', 'gender', 'race_ethnicity']]

data = data.dropna(subset=['grade_level'])
targets = targets[data.index]  # Align targets with data after dropping NaNs

# One-hot encode the categorical columns
data_encoded = pd.get_dummies(data, columns=['gender', 'race_ethnicity'])

# Now data_encoded can be used for logistic regression
print(data_encoded.head())
```

```python
train_x, test_x, train_y, test_y = sklearn.model_selection.train_test_split(data_encoded, targets, random_state=3)

# Create logistic regression model
model = sklearn.linear_model.LogisticRegression()
model.fit(train_x, train_y)
```

```python
# Make predictions
y_hat = model.predict(test_x)

# y_hat is the estimation
def check_predictions(y_hat, y):
    return np.sum(y_hat == y) / len(y_hat)

print("Predictions: ")
print(y_hat)

print(f"Accuracy: {check_predictions(y_hat, test_y) * 100:.2f}%")
```
